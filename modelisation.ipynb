```python
from sklearn.model_selection import train_test_split

def split_data(df, target_column='target', test_size=0.2, random_state=42):
    """
    Paramètres :
    - df : DataFrame, le DataFrame contenant les données.
    - target_column : str, le nom de la colonne cible.
    - test_size : float, la proportion des données à inclure dans l'ensemble de test.
    - random_state : int, la graine pour la reproductibilité des résultats.

    Retourne :
    - X_train, X_test, y_train, y_test : les ensembles d'entraînement et de test.
    """
    X = df.drop(target_column, axis=1)
    y = df[target_column]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    return X_train, X_test, y_train, y_test

# Utilisation de la fonction
X_train, X_test, y_train, y_test = split_data(train_df_sample)
```
