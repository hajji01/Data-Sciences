```python
import pandas as pd
import sklearn
import numpy as np
```

```python
train_df_sample = pd.read_csv("./train_df_sample.csv")
train_df_sample
```

```python

train_df_sample = train_df_sample[train_df_sample["target"].isna() == False]
```

```python
from sklearn.model_selection import train_test_split

X = train_df_sample.drop('target', axis=1)
y = train_df_sample['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Supprimer la colonne datetime ou la transformer en format numérique
# Vous pouvez choisir l'une des deux options suivantes en fonction de vos besoins
X_train = X_train.drop('datetime', axis=1)  # Supprimer la colonne datetime
# OU
# X_train['datetime'] = pd.to_datetime(X_train['datetime']).astype(int)  # Transformer en format numérique

# Entraîner le modèle
reg = LinearRegression().fit(X_train, y_train)

# Prédire sur les données d'entraînement
y_train_hat = reg.predict(X_train)

# Calculer l'erreur quadratique moyenne sur les données d'entraînement
mse_train = mean_squared_error(y_train, y_train_hat)
print("MSE on training data:", mse_train)
```
